```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/preprocessed/Smart_Farming_Crop_Yield_2024.csv")

num_features = ["soil_moisture_%", "soil_pH", "pesticide_usage_ml", "yield_kg_per_hectare"] #check the data
df[num_features].hist(figsize=(10,8), bins=20)
plt.suptitle("Распределение числовых признаков")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[num_features])
plt.title("Выбросы в числовых признаках")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_features].corr(), annot=True, cmap="coolwarm")
plt.title("Корреляция числовых признаков")
plt.show()
```

```python
cat_features = ["region_le", "croptype_le", "fertilizer_type_le", "irrigation_type_le", "cropdisease_le"]
for col in cat_features:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col])
    plt.title(f"Распределение категориального признака: {col}")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["longitude"], y=df["latitude"], hue=df["yield_kg_per_hectare"], palette="viridis")
plt.title("География урожайности")
plt.show()
```
